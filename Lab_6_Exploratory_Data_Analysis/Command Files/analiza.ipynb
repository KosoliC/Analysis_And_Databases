```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
df = pd.read_csv('../Analysis Data/converted_data.csv')
df.head(10)
```

## Plik zawiera analize każdej ze zmiennych, oraz zdefiniowane watości bins, dla zmiennych "Wiek kupującego" oraz "Dni od zakupu", aby je odpowiednio zebrać w przedziały dla lepszej wizualizacji na wykresach. 

```python
df['Marka'].value_counts().plot.bar()
plt.title('Podział kupionych odkurzaczy ze względu na markę.')
plt.ylabel('Ilość zakupionych odkurzaczy danej marki')
plt.xlabel('Rodzaje marek produkujących odkurzacze')
plt.xticks(rotation=45)
plt.savefig(r'..\Documents\data_appendix_bar1_brand')
plt.show()

```

```python
# zdefiniowanie bins dla zmiennej "Dni o zakupu"
bins_dni_od_zakupu =  [0,5,9,13,18]
labels_dni_od_zakupu = ['1-4 dni','5-8 dni', '9-12 dni', 'ponad 13 dni']
df['Dni od zakupu'] = pd.cut(x = df['Dni od zakupu'], bins = bins_dni_od_zakupu, labels = labels_dni_od_zakupu, include_lowest = True)

df['Dni od zakupu'].value_counts().sort_index().plot.bar()
plt.title('Ilość dni od zakupu po jakich wystawiono ocenę')
plt.ylabel('Liczba wystawionych ocen po określonej liczbie dni')
plt.xlabel('Ilość dni która mineła od zakupu')
plt.xticks(rotation=45)
plt.savefig(r'..\Documents\data_appendix_bar2_time_of_the_assessment')
plt.show()
```

```python
# zdefiniowanie bins dla zmiennej "Wiek kupującego"
bins_wiek_kupujących =  [18,31,41,51,61,100]
labels_wiek_kupujących = ['18-30 lat','31-40 lat', '41-50 lat', '51-60 lat','ponad 60 lat']
df['Wiek kupującego'] = pd.cut(x = df['Wiek kupującego'], bins = bins_wiek_kupujących, labels = labels_wiek_kupujących, include_lowest = True)

df['Wiek kupującego'].value_counts().sort_index().plot.bar()
plt.title('Histogram ilości osób w danym wieku która dokonała zakupu')
plt.ylabel('Ilość osób w danym wieku')
plt.xlabel('Wiek w latach')
plt.xticks(rotation=45)
plt.savefig(r'..\Documents\data_appendix_bar3_age')
plt.show()
```

```python
df['Płeć kupującego'].value_counts().plot.bar()
plt.title('Podział kupujących ze względu na płeć')
plt.ylabel('Ilość zakupionych odkurzaczy')
plt.xticks(rotation=45)
plt.savefig(r'..\Documents\data_appendix_bar4_sex')
plt.show()
```

```python
brand_rating = df.groupby('Marka')['Ocena']
mean_brand_rating = brand_rating.mean()
ax = mean_brand_rating.plot.bar()
plt.xticks(rotation=45)
ax.bar_label(ax.containers[0])
plt.title('Średnia ocen odkurzaczy dla danych marek')
plt.ylabel('Średnia ocena')
plt.xlabel('Rodzaje marek produkujących odkurzacze')
plt.savefig(r'..\Documents\data_appendix_bar5_mean')
plt.show()
```
