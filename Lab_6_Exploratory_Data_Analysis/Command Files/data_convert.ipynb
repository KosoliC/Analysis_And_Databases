```python
import pandas as pd

data = pd.read_csv('../Original Data/10_POMORSKIE.csv')
# Czyszczenie danych

data = data.drop(data.columns[[0]], axis=1)
data['Płeć kupującego'] = data['Płeć kupującego'].replace(['K','M','bd.'],['Kobieta','Mężczyzna','Brak danych o płci'])
data['Wiek kupującego'] = data['Wiek kupującego'].astype('Int64')
data.head(15)
```

```python
# Zapis wyczysczonych danych do folderu Analysis Data
data.to_csv(r'..\Analysis Data\converted_data.csv',index=False)
```
